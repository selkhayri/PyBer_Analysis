```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

### Create a line chart using the MATLAB method

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

### Create a Line Chart Using the Object-Oriented Interface

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

### The fig, ax = plt.subplots() line of code can also be written like this.

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

### To create a quick graph using the object-oriented interface method, use the following two lines

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

### Use plt.show()

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

## 5.1.4 Annotate Charts

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

## 5.1.7 Create Scatter Plots and Bubble Charts

### Skill Drill

Using the object-oriented approach, make the following changes to your bubble chart:

    Change the color of the markers to sky blue.
    Change the size of the markers to 5 times each data point.
    Make the color 20% transparent.
    Add a black edge color to the circles.
    Make the linewidth of the circles 2 points.
    Add a legend for the city of Boston.
    Add a title.
    Switch the axis so that the Fare($) data is on the x-axis.
    Add a limit for the x-axis data.
    Invert the y-axis so that January is at the top. 

```python
fig, ax = plt.subplots()

# Change the color of the markers to sky blue.
color = "skyblue"

# Change the size of the markers to 5 times each data point.
y_axis_5 = [ s * 5 for s in y_axis]

# Make the color 20% transparent.
alpha=0.8

# Add a black edge color to the circles.
edgecolors = ["black"]

# Make the linewidth of the circles 2 points.
linewidths = [2]

# Switch the axis so that the Fare($) data is on the x-axis.

ax.scatter(y_axis, x_axis, s=y_axis_5, c=color,alpha=alpha,edgecolors=edgecolors, \
          linewidths = linewidths)

# Add a legend for the city of Boston.
legend = ["Boston"]
ax.legend(legend)

# Add a title
ax.set_title("PyBer Fare by Month")

# Add a limit for the x-axis data.
ax.set_xlim(0,50)

# nvert the y-axis so that January is at the top. 
ax.invert_yaxis()
```

## 5.1.8 Create Pie Charts

### Create a Pie Chart Using the MATLAB Method

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# Increase the size of the pie chart
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

### Skill Drill

Using the object-oriented approach, make the following changes to the pie chart:

    Add a percentage to one decimal place to each wedge of the pie.
    Increase the figure size to 8x8.
    Explode the two highest percentage months.
    Add a shadow.
    Add a start angle so that January is at the top.
    Reverse the order so that the month order is in a clockwise direction.
    Add new colors of your own choosing or use the colors from the previous pie chart.

```python
# Add a percentage to one decimal place to each wedge of the pie.
autopct = "%.1f%%"

# Increase the figure size to 8x8.
figsize = (8,8)
fig, ax = plt.subplots(figsize=figsize)

# Explode the two highest percentage months.
explode_values = [0,0,0.2,0,0,0,0.2,0,0,0,0,0]

# Add a shadow.
shadow = True

# Add a start angle so that January is at the top.
startangle = 90

# Reverse the order so that the month order is in a clockwise direction.
counterclockwise = False

# Add new colors of your own choosing or use the colors from the previous pie chart.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# Draw the pie chart
ax.pie(y_axis,labels=x_axis,\
       autopct=autopct,\
       explode=explode_values,\
       shadow=shadow,\
       startangle=startangle,\
       counterclock=counterclockwise,\
       colors=colors)

plt.show()

```
